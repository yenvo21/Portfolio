```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
 df= pd.read_csv('degrees-that-pay-back.csv')
```

```python
# Rename cols.
df.columns = ['major', 'start', 'mid_career', 'delta_start_mid', 'mid_10p',
            'mid_25p', 'mid_75p', 'mid_90p']

```

```python
# Check dtypes for salary figures.
type(df.start[0])
```

```python
salary_cols = ['start', 'mid_career', 'mid_10p', 'mid_25p', 'mid_75p', 'mid_90p']

for col in salary_cols:
    df[col] = df[col].str.replace('$', '') # remove dollar signs.
    df[col] = df[col].str.replace(',', '').astype(float) # remove commas and convert to floats.
    
df.head()
```

```python
df.isnull().any()
```

```python
df.duplicated().any()
```

```python
df.describe()
```

```python
# Bar plot for starting salaries by major.
plt.figure(figsize = (10, 10))
sns.barplot(x = df.sort_values('start', ascending = False).start, # Put data in descending order by start salary.
            y = df.sort_values('start', ascending = False).major) # Makes the graph easier to interpret.
plt.title('Mean Starting Salaries\nby College Major', fontsize = 12)
plt.xlabel('Salary (USD)', fontsize = 14)
plt.ylabel('Major', fontsize = 14)
plt.grid(axis = 'x')
```

```python
df1 = pd.read_csv('salaries-by-college-type.csv')
```

```python
df1.head()
```

```python
df.isnull().any()
```

```python
df.duplicated().any()
```

```python
#Some values are missing
#drop the values columns that have missing data 
df1.dropna(inplace=True)
df1.reset_index(inplace=True, drop=True)
df1.head()
```

```python
#Change strings to floats using str.replace
#Only replace strings in columns that are numeric (will not work eitherwise)
for x in df1.columns:
    if x != 'School Name' and x != 'School Type':
        new_col = df1[x].str.replace("$", "")
        new_col  = new_col .str.replace(",", "")
        df1[x] = pd.to_numeric(new_col )
```

```python
df1.head()
```

```python
#To view the number of schools by type:
df1["School Type"].value_counts()
```

```python
#Plot school type by Starting Median Salary
sns.FacetGrid(df1, hue="School Type", size=5) \
   .map(sns.kdeplot, "Starting Median Salary") \
   .add_legend()
plt.show()
```

```python
#Use a heatmap to view the correlation between variables in the dataset.

corr = df1.corr()
corr = (corr)
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values,cmap="YlGnBu")
        
plt.show()
```

```python
#A bar plot of the average Mid-Career Salary by school type
objects = ('Engineering', 'Party',' Liberal Arts' ,'Ivy League','State')
y_pos = np.arange(len(objects))
performance = [103842,84685,89379,120125,78567]
 
plt.bar(y_pos, performance, align='center', alpha=0.5,  color='r')
plt.xticks(y_pos, objects, rotation='70')
plt.ylabel('Average')
plt.title('Average Mid-Career Salary by School Type')
 
plt.show()
```

```python
#Salaries group by school Type
df1.groupby('School Type').mean().plot(kind = 'bar', figsize = (10,7))
```
